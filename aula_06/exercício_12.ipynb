### Exercício 12 - Inner Join
#### Objetivo: Unir os DataFrames de livros e empréstimos usando o método inner join e mostrar apenas os livros que foram emprestados. 
#### Instruções: 
#### 1.Carregue os arquivos livros.csv e emprestimos.csv em DataFrames separados. 
#### 2. Realize um inner join entre as tabelas, unindo-as pela coluna id_livro.
#### 3. Exiba o resultado, mostrando as colunas: titulo, autor, nome_leitor, data_emprestimo e data_devolucao.

```python
import pandas as pd
```

```python
livros = pd.read_csv("livros.csv")
emprestimos = pd.read_csv("emprestimos.csv")
```

```python
livros_emprestados = pd.merge(livros, emprestimos, on='id_livro', how='inner')
livros_emprestados
```

```python
resultado = livros_emprestados[['titulo', 'autor', 'nome_leitor', 'data_emprestimo', 'data_devolucao']]
resultado
```
