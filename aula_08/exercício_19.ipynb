```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('palmerpenguins_extended.csv')
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='bill_length_mm', y='flipper_length_mm', hue='species', style='species')
plt.title('Comprimento do Bico vs Comprimento da Asa')
plt.xlabel('Comprimento do Bico (mm)')
plt.ylabel('Comprimento da Asa (mm)')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(x= 'species', y= 'body_mass_g', data=df)
plt.title('Comparação de Peso entre as Espécies de Pinguins')
plt.xlabel('Espécie')
plt.ylabel('Peso (gramas)')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.histplot(data=df, x='bill_length_mm', hue='species', kde= True)
plt.title('Distribuição do Comprimento do Bico entre as Espécies')
plt.xlabel('Comprimento do Bico (mm)')
plt.ylabel('Frequência')
plt.show()
```

#### À medida que as mudanças climáticas continuam a impactar os ecossistemas polares, as espécies de pinguins enfrentam desafios diferentes em sua adaptação. Analisando os dados das três espécies principais – Adelie, Gentoo e Chinstrap – notamos que os pinguins Gentoo, com suas asas mais longas , estar melhor equipado para enfrentar o aumento das áreas de água abertas, resultado da diminuição do gelo marinho. 

#### Enquanto isso, o peso dos pinguins Adelie parece ter diminuído, revelando uma escassez possível de alimentos à medida que as águas aquecem e alteram a cadeia. alimentar local. A variabilidade no comprimento do bico entre as espécies também pode refletir a adaptação a novas fontes de alimento Em conjunto, esses dados ilustram como essas aves marinhas estão reagindo a um ambiente em rápida transformação.
