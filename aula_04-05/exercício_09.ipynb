```python
import pandas as pd 
```

```python
filmes = ['Como perder um homem em 10 dias', 'Trainspotting', 'Mid90', 'Buffalo 66', 'Sobre lobos e meninos']
ano_lancamento = [2003, 1996, 2018, 1998, 2003]

dadosfilmes = pd.DataFrame ({
    'Filmes' : filmes,
    'Lançamento' : ano_lancamento
})
notas_filmes = [10,9,9,10,10]

dadosfilmes['Notas'] = notas_filmes

series = ['Breaking Bad', 'Friends', 'Dark', 'Peaking Blinders', 'Hannibal']
ano_lancamento = [2008, 1994, 2017, 2013, 2013]
notas_series = [10, 10, 10, 10, 9]
 
dadosseries = pd.DataFrame ({
    'Séries' : series,
    'Lançamento' : ano_lancamento,
    'Notas' : notas_series
})
desenhos = ['Coraline', 'Por Água Abaixo', 'Shrek', 'A Noiva Cadáver', 'Os Sem Florestas']
ano_lancamento = [2009, 2006, 2001, 2005, 2006]
notas_desenhos = [10, 9, 10, 10, 9]

dadosdesenhos = pd.DataFrame ({
    'Desenhos' : desenhos,
    'Lançamento' : ano_lancamento,
    'Notas' : notas_desenhos
})
```

```python
filme_maior_nota = dadosfilmes[dadosfilmes['Notas'] == dadosfilmes['Notas'].max()]
filme_maior_nota
```

```python
serie_maior_nota = dadosseries[dadosseries['Notas'] == dadosseries['Notas'].max()]
serie_maior_nota
```

```python
desenho_maior_nota = dadosdesenhos[dadosdesenhos['Notas'] == dadosdesenhos['Notas'].max()]
desenho_maior_nota
```

```python
dados_todos = pd.concat([dadosfilmes, dadosseries, dadosdesenhos], keys=['Filme', 'Série', 'Desenho'])
dados_todos.reset_index(level=0, inplace=True)
dados_todos.rename(columns={'level_0': 'Tipo'}, inplace=True)

dados_todos.to_json('dados_combinados.json', orient='records')
```
