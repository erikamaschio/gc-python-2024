### Exercício 11 – Mini Boss - Coop
#### ✦ Crie e exiba um dataframe musicas com informações sobre suas músicas favoritas ( 5 músicas ):
#### ✦ id_musica – identificador da música.
#### ✦ nome_musica – nome da música.
#### ✦ artista – nome do artista ou banda
#### ✦ genero – gênero musical (Pop, Rock, Jazz, etc.)
#### ✦ duracao – duração da música em minutos.

```python
import pandas as pd
```

```python
dados_musicas = {
    'id_musica': [1, 2, 3, 4, 5],
    'nome_musica': ['Jesus Chorou', 'Um bom lugar', 'Favela Sinistra', 'Bohemian Rhapsody', 'Bornie to Die'],
    'artista': ['Racionais', 'Sabotage', 'Trilha Sonora do Gueto', 'Queen', 'Lana del Rey'],
    'genero': ['Rap','Rap', 'Rap', 'Rock', 'Alternativa/Indie' ],
    'duracao': [07.52, 05.39, 03.36, 05.54, 04.47]
}

df_musicas = pd.DataFrame(dados_musicas)
df_musicas
```

```python
dados_reproducoes = {
    'id_reproducao': [101, 102, 103, 104, 105],
    'id_musica': [1, 2, 3, 4, 5],
    'data_reproducao': ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05'],
    'duracao_reproducao': [3.03, 5.55, 7.30, 4.00, 6.30] 
}

df_reproducoes = pd.DataFrame(dados_reproducoes)
df_reproducoes
```

```python
# Simulando novas 5 reproduções com diferentes músicas
novas_reproducoes = {
    'id_reproducao': [106, 107, 108, 109, 110],
    'id_musica': [2, 3, 1, 5, 4],
    'data_reproducao': ['2024-10-06', '2024-10-07', '2024-10-08', '2024-10-09', '2024-10-10'],
    'duracao_reproducao': [5.00, 6.50, 3.03, 6.00, 4.54]  # Algumas reproduções podem ser parciais
}

# Adicionando as novas reproduções ao dataframe original de reproduções usando concat
df_reproducoes = pd.concat([df_reproducoes, pd.DataFrame(novas_reproducoes)], ignore_index=True)

# Realizando o merge entre os dataframes musicas e reproducoes pela coluna id_musica
df_merged = pd.merge(df_reproducoes, df_musicas, on='id_musica')

# Selecionando e exibindo nome da música, artista e duração reproduzida
df_resultado = df_merged[['nome_musica', 'artista', 'duracao_reproducao']]
df_resultado
```
